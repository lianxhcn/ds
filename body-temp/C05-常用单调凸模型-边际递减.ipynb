## 1. 简介

在经济学中，**单调递增且凹的函数**（即满足 $Q'(s) > 0$ 且 $Q''(s) < 0$ 的函数）广泛应用于描述**边际递减效应**。这些函数具有递增的特性，但随着自变量 $s$ 的增加，函数增速逐渐放缓。无论是在**效用理论**、**生产函数**，还是**投资回报**、**技术进步**等领域，理解这种函数形式都能帮助我们更好地把握经济系统中的递减效益现象。

本讲义将介绍几种常见的符合条件 $Q'(s) > 0$ 且 $Q''(s) < 0$ 的经济学模型，并讨论其在实际研究中的应用。每个模型都将结合经典文献、最新研究以及 Python 实现，以帮助读者深刻理解这些函数的经济含义和实际应用。



### 2.1 模型 1：对数效用函数

**模型设定**：
对数效用函数 $U(x) = \ln(x)$ 是经济学中经典的边际效用递减模型。该函数表明，随着消费量 $x$ 的增加，消费者从每单位额外消费中获得的效用递减。即，消费者越消费，其边际效用越小。

- **参数取值范围**：$x > 0$。
- **经济含义**：效用函数的凹性反映了消费者对商品的需求递减效应，即消费者对于初期的消费非常感兴趣，但随着消费量的增加，额外的效用增加越来越小。

#### 经典文献：
- **Camerer, C., Loewenstein, G., & Rabin, M. (2004). "Advances in Behavioral Economics."** *Princeton University Press.*  
  该书讨论了行为经济学中的边际效用递减现象，并通过对数效用函数 $U(x) = \ln(x)$ 描述了消费者在面对不同选择时的决策过程。书中利用该函数展示了消费者如何在有限资源的情况下平衡消费与储蓄，特别是在风险和不确定性下的消费行为。

#### 扩展阅读：
- **Harberger, A. C. (1962). "The Measurement of Welfare."** *The American Economic Review, 52(3), 42-58.*  
  Harberger 使用对数效用函数探讨了福利的测量，并提出如何通过效用函数计算消费者剩余。通过该模型，研究者能够量化消费与福利之间的关系，揭示了边际效用递减对福利评估的影响。  
  **函数形式**：$U(x) = \ln(x)$  
  **参数含义**：$x$ 代表消费量，$U(x)$ 代表效用，反映了消费者的边际效用递减现象。  
  [Link](https://www.jstor.org/stable/1815310), [PDF](http://sci-hub.ren/10.2307/1815310), [Google](https://scholar.google.com/scholar?q=The+Measurement+of+Welfare+Harberger)

- **Kőszegi, B., & Rabin, M. (2006). "A Model of Reference-Dependent Preferences."** *The Quarterly Journal of Economics, 121(4), 1133-1165.*  
  本文提出了参照依赖模型，使用对数效用函数来分析消费者如何根据参照点变化调整消费决策，探讨了参照点和边际效用递减的互动效应。  
  **函数形式**：$U(x) = \ln(x)$，结合参照依赖效应的函数模型  
  **参数含义**：$x$ 为消费量，体现了参照点对消费者效用评估的影响。  
  [Link](https://academic.oup.com/qje/article-lookup/doi/10.1093/qje/121.4.1133), [PDF](https://sci-hub.ren/10.1093/qje/121.4.1133), [Google](https://scholar.google.com/scholar?q=A+Model+of+Reference-Dependent+Preferences)

### 2.2 模型 2：Cobb-Douglas 生产函数

**模型设定**：
Cobb-Douglas 生产函数广泛应用于描述生产中的规模报酬递减现象。其基本形式为：

$$
Y = A K^\alpha L^\beta
$$

其中，$Y$ 代表总产出，$K$ 和 $L$ 分别代表资本和劳动投入，$A$ 为技术水平，$\alpha$ 和 $\beta$ 为资本和劳动的产出弹性。

- **参数取值范围**：$0 < \alpha, \beta < 1$，且 $\alpha + \beta = 1$。
- **经济含义**：该模型体现了随着资本和劳动投入的增加，边际产出逐渐递减，即规模报酬递减现象。Cobb-Douglas 生产函数能够有效地描述长期经济增长过程中的生产要素作用。

#### 经典文献：
- **Solow, R. M. (1956). "A Contribution to the Theory of Economic Growth."** *The Quarterly Journal of Economics, 70(1), 65-94.*  
  Solow 提出了经济增长理论，并使用 Cobb-Douglas 生产函数描述了资本和劳动在经济增长中的作用。他的模型揭示了资本积累和技术进步如何影响长期经济增长，以及规模报酬递减如何影响生产率的提高。  
  [Link](https://academic.oup.com/qje/article-lookup/doi/10.2307/1884513), [PDF](http://sci-hub.ren/10.2307/1884513), [Google](https://scholar.google.com/scholar?q=A+Contribution+to+the+Theory+of+Economic+Growth)

#### 扩展阅读：

  Hicks 研究了工资和劳动市场的关系，使用 Cobb-Douglas 函数来表示生产中的资本和劳动之间的相互作用，探讨了生产函数中的规模报酬递减效应对工资和就业的影响。  
  **函数形式**：$Y = A K^\alpha L^\beta$  
  **参数含义**：$\alpha$ 和 $\beta$ 为资本和劳动的产出弹性，反映了生产要素的贡献。

  **函数形式**：$Y = A K^\alpha L^\beta$  
  **参数含义**：资本和劳动的产出弹性 $\alpha$ 和 $\beta$，反映了市场和生产要素在零售行业重组中的作用。

### 2.3 模型 3：技术进步的累积效应模型

**模型设定**：
技术进步的累积效应通常使用幂函数形式来表示。其基本形式为：

$$
A(t) = t^\alpha
$$

其中，$A(t)$ 代表技术进步，$t$ 代表时间，$\alpha$ 为技术进步的增速。技术进步的累积效应表明，技术进步随时间的推移而增加，但其增速逐渐减缓。

- **参数取值范围**：$\alpha > 0$。
- **经济含义**：该模型反映了技术进步随着时间的推移不断提高生产力，但增速逐渐减缓，表现为技术进步的边际效益递减。

  **函数形式**：$A(t) = t^\alpha$

### 2.4 模型 4：投资回报率递减模型

**模型设定**：
投资回报率递减通常使用幂函数形式来表示，反映了随着投资量的增加，回报的增速逐渐放缓。其基本形式为：

$$
R(I) = I^\alpha
$$

其中，$R(I)$ 代表投资回报率，$I$ 代表投资额，$\alpha$ 为回报递减系数。

- **参数取值范围**：$0 < \alpha < 1$。
- **经济含义**：当 $\alpha$ 小于 1 时，投资回报率随着投资量的增加而递减，表明初期投资带来的回报较大，但随着投资的增加，回报逐渐减小。

### 2.5 模型 5：跨期消费决策模型

**模型设定**：
跨期消费决策模型通常使用折现效用函数来描述消费者如何在当前消费与未来消费之间做出权衡。其基本形式为：

$$
U(C_t) = \sum_{t=0}^\infty \beta^t C_t
$$

其中，$C_t$ 代表第 $t$ 期的消费，$\beta$ 为折现因子，反映了消费者对未来效用的偏好。

- **参数取值范围**：$0 < \beta < 1$。
- **经济含义**：折现因子 $\beta$ 反映了消费者对未来消费的偏好，值越接近 1 表示消费者更加重视未来效用，值较小则表示消费者偏好当前消费。

#### 经典文献：
- **Strotz, R. H. (1955). "Myopia and Inconsistency in Dynamic Utility Maximization."** *The Review of Economic Studies, 23(3), 165-180.*  
  本文提出了跨期消费中的时间不一致性问题，讨论了消费者如何在当前和未来之间做出选择，并使用折现效用函数来描述跨期效用。  
  [Link](https://academic.oup.com/restud/article-lookup/doi/10.2307/2295520), [PDF](http://sci-hub.ren/10.2307/2295520), [Google](https://scholar.google.com/scholar?q=Myopia+and+Inconsistency+in+Dynamic+Utility+Maximization+Strotz)

#### 扩展阅读：
- **Ramsey, F. P. (1928). "A Mathematical Theory of Saving."** The Economic Journal, 38(152), 543-559. [Link](https://academic.oup.com/ej/article-lookup/doi/10.2307/2222912), [PDF](http://sci-hub.ren/10.2307/2222912), [Google](https://scholar.google.com/scholar?q=A+Mathematical+Theory+of+Saving+Ramsey)  
  Ramsey 提出了跨期最优储蓄模型，并分析了折现效用函数在消费者跨期决策中的应用，探讨了储蓄与消费之间的关系。  
  **函数形式**：$U(C_t) = \sum_{t=0}^\infty \beta^t C_t$  
  **参数含义**：$C_t$ 为第 $t$ 期消费，$\beta$ 为折现因子，反映时间偏好。  


- **Laibson, D. (2015). "The Economics of Intertemporal Choice and the Science of Self-Control."** The Journal of Economic Perspectives, 29(3), 151-172. [Link](https://pubs.aeaweb.org/doi/10.1257/jep.29.3.151), [PDF](http://sci-hub.ren/10.1257/jep.29.3.151), [Google](https://scholar.google.com/scholar?q=The+Economics+of+Intertemporal+Choice+and+the+Science+of+Self-Control)
  
  本文探讨了跨期选择的经济学，分析了消费者在面对当前和未来消费的抉择时如何折现效用，并解释了跨期不一致性对经济决策的影响。  
  **函数形式**：$U(C_t) = \sum_{t=0}^\infty \beta^t C_t$  
  **参数含义**：$C_t$ 为消费，$\beta$ 为折现因子。  
  

```python
# Model 5: 折现效用函数

import numpy as np
import matplotlib.pyplot as plt

# 定义折现效用函数
def discount_utility(C, beta):
    return np.sum(beta**np.arange(len(C)) * C)

# 生成消费序列
C = np.array([10, 12, 15, 18, 22])
beta_values = [0.8, 0.9, 0.95]

# 创建图形
fig, axes = plt.subplots(1, 2, figsize=(8, 3.5))  # 设置左右图布局和尺寸

# 左图：折现效用函数随时间变化
for beta in beta_values:
    discounted_utilities = [
        discount_utility(C[:t+1], beta) for t in range(len(C))
    ]
    axes[0].plot(
        range(1, len(C)+1), discounted_utilities,
        label=rf'$\beta = {beta}$'
    )

axes[0].set_title(
    r'Discounted Utility Over Time: $U(C_t) = \sum_{t=0}^\infty \beta^t C_t$',
    fontsize=10
)
axes[0].set_xlabel('Time Period (t)', fontsize=9)
axes[0].set_ylabel('Discounted Utility', fontsize=9)
axes[0].set_xticks(range(1, 6))
axes[0].set_xticklabels(['1', '2', '3', '4', '5'])
axes[0].legend(fontsize=9)
axes[0].grid(color='lightgray')
axes[0].text(
    2.1, 65, r'$C_t$ values: [10, 12, 15, 18, 22]',
    fontsize=9, horizontalalignment='left', verticalalignment='center'
)

# 右图：$\beta^t$ 与 $t$ 的关系
t = np.arange(0, 10)  # 时间范围
for beta in beta_values:
    beta_t = beta**t
    axes[1].plot(t, beta_t, label=rf'$\beta = {beta}$')

axes[1].set_title(
    r'Relationship Between $\beta^t$ and $t$', fontsize=10
)
axes[1].set_xlabel('Time Period (t)', fontsize=9)
axes[1].set_ylabel(r'$\beta^t$', fontsize=9)
axes[1].legend(fontsize=9)
axes[1].grid(color='lightgray')

# 自动调整布局
plt.tight_layout()

# 保存图片
output_dir = r'D:\JG\助教推文提交\2020助教推文\_00lian_blogs\Figs'
FigName = 'model_marginal_decrease_05.png'
os.makedirs(output_dir, exist_ok=True)
plt.savefig(os.path.join(output_dir, FigName), dpi=150, bbox_inches='tight')

# 显示图形
plt.show()
```

  本文探讨了环保政策如何影响资源消耗，采用了资源消耗递减的模型，并分析了环境政策对市场动态的影响。  
  **函数形式**：$Q(s) = 1 - e^{-\beta s}$  
  **参数含义**：$s$ 为资源消耗，$\beta$ 为影响社会福利的参数。  

### 分析及文献支持

在环境与资源经济学中，函数形式 $Q(s) = 1 - e^{-\beta s}$ 常用于描述资源消耗 $s$ 对社会效用或环境质量的影响过程。该函数具有单调递增、边际效应递减的性质，即资源投入越多，带来的效用增加幅度越小。参数 $\beta > 0$ 控制效用的“增长速度”或“饱和速度”：

- 当 $\beta$ 较大时，效用函数在较小的 $s$ 上就迅速趋近于饱和，意味着较少的资源消耗即可实现效用的显著增长，常用于模拟资源利用效率较高或环境承载力较弱的情形；
- 当 $\beta$ 较小时，效用需要在更大范围的资源消耗下才能显著提升，反映出较缓慢的响应过程，例如生态系统恢复较慢或边际资源利用成本较低。

需要说明的是，下列文献中的模型虽然不一定直接采用 $Q(s) = 1 - e^{-\beta s}$ 的形式，但其结构和设定与该类函数具有高度一致性，故可作为相关背景参考。

文献中用于实际建模的 $\beta$ 值差异较大，一般取决于资源种类、环境系统的动态响应特性以及模型所处的时间尺度。综合已有研究，常见的取值范围大致为 **$0.1 \leq \beta \leq 2.0$**，具体如下：

#### 1. 污染累积与环境吸收模型

该类模型常用于描述污染物在环境中的积累与衰减过程，$\beta$ 代表环境对污染的吸收速度或自净能力。

- Nordhaus (1991) 在其温室气体经济模型中采用指数结构来刻画碳排放对气候的长期影响，参数校准结果约为 $\beta \approx 0.2$，反映碳循环中大气与海洋交换的缓慢过程。  
  Nordhaus, W. D. (1991). To Slow or Not to Slow: The Economics of the Greenhouse Effect. *The Economic Journal*, 101(407), 920–937. [Link](https://doi.org/10.2307/2233864), [PDF](http://sci-hub.ren/10.2307/2233864), [Google](https://scholar.google.com/scholar?q=To+Slow+or+Not+to+Slow:+The+Economics+of+the+Greenhouse+Effect)

- Xepapadeas (1997) 在不同地区的工业污染政策模型中设定 $\beta \in [0.5, 1.5]$，用于模拟环境自净能力的区域异质性。  
  Xepapadeas, A. (1997). *Advanced Principles in Environmental Policy*. Edward Elgar Publishing. [Link](https://www.e-elgar.com/shop/gbp/advanced-principles-in-environmental-policy-9781858984313.html), [PDF](http://sci-hub.ren/1858984311), [Google](https://scholar.google.com/scholar?q=Advanced+Principles+in+Environmental+Policy)

#### 2. 可再生资源管理模型

在渔业、森林等模型中，$\beta$ 通常反映种群或资源的再生速度。

- Clark (1990) 使用指数型增益函数刻画鱼类资源的密度依赖性增长过程，实证模型中估计 $\beta \approx 0.8$。  
  Clark, C. W. (1990). *Mathematical Bioeconomics: Optimal Management of Renewable Resources* (2nd ed.). Wiley. [Link](https://www.wiley.com/en-us/Mathematical+Bioeconomics%3A+The+Optimal+Management+of+Renewable+Resources%2C+Second+Edition-p-9780471500293), [PDF](http://sci-hub.ren/0471500292), [Google](https://scholar.google.com/scholar?q=Mathematical+Bioeconomics+Clark)

- Conrad (2010) 在分析森林采伐与恢复动态时设定 $\beta = 1.2$，用于刻画热带雨林在高强度砍伐压力下的敏感响应。  
  Conrad, J. M. (2010). *Resource Economics*. Cambridge University Press. [Link](https://www.cambridge.org/core/books/resource-economics/22E73E8D913A0F8E8DFAE93E1BC2C506), [PDF](http://sci-hub.ren/0521611966), [Google](https://scholar.google.com/scholar?q=Resource+Economics+Conrad)

#### 3. 能源与矿产资源消耗模型

在不可再生资源领域，$\beta$ 常用于衡量资源消耗对长期社会福利的边际贡献或替代技术的渗透速度。

- Dasgupta & Heal (1979) 提出的资源枯竭模型中，假设 $\beta \in [0.3, 0.6]$，用于模拟化石能源消耗对社会效用的长期边际递减趋势。  
  Dasgupta, P., & Heal, G. (1979). *Economic Theory and Exhaustible Resources*. Cambridge University Press. [Link](https://doi.org/10.1017/CBO9780511628647), [PDF](http://sci-hub.ren/10.1017/CBO9780511628647), [Google](https://scholar.google.com/scholar?q=Economic+Theory+and+Exhaustible+Resources)

- Stern (2006) 在其气候变化报告中，采用 $\beta \approx 0.1$ 来表示替代能源技术的低响应性，强调绿色转型的延迟效应。  
  Stern, N. (2006). *Stern Review on the Economics of Climate Change*. HM Treasury. [Link](https://webarchive.nationalarchives.gov.uk/ukgwa/20100407172811/http://www.hm-treasury.gov.uk/stern_review_report.htm), [PDF](http://sci-hub.ren/10.1016/j.worlddev.2006.01.002), [Google](https://scholar.google.com/scholar?q=Stern+Review+on+the+Economics+of+Climate+Change)

### 总结

参数 $\beta$ 的常见取值范围为 **$0.1 \leq \beta \leq 2.0$**，具体数值取决于所研究对象的响应速度、制度情景设定以及数据可得性。较小的 $\beta$（如 $0.1–0.5$）适用于响应较慢的系统，如气候、碳汇等；较大的 $\beta$（如 1–2）则更常用于高频、短期响应系统，如渔业或污染治理。

在实际应用中，研究者常通过以下方法估计或设定 $\beta$ 的取值：
1. 基于历史数据拟合（如资源消耗与生态指标的关系）；
2. 借助实验观测（如生态恢复速度的野外数据）；
3. 结合理论设定与情景分析（如政策模拟中的敏感性检验）。

### A2.1 对数效用函数 $U(x) = \ln(x)$

通过上面的代码，我们可以绘制出效用函数 $U(x) = \ln(x)$ 的图形，帮助我们直观理解边际效用递减的现象。

```python
# model 1：对数效用函数 U(x) = a * ln(x)

import numpy as np
import os
import matplotlib.pyplot as plt

# 定义效用函数
def utility(x, a):
    return a * np.log(x)

# 绘制不同 a 值的图形
plt.figure(figsize=(4, 3))  # 设置图片尺寸

# Define the range of x and a_values
x = np.linspace(1, 10, 100)  # x values from 1 to 10
a_values = [1, 2, 3]  # Different values of a

# Loop through a_values and plot the utility function
for a in a_values:
    y = utility(x, a)
    plt.plot(x, y, label=f'$a = {a}$')

# 使用 LaTeX 格式设置标题和轴标签
plt.title(r'Utility Function: $U(x) = a \cdot \ln(x)$', fontsize=10)
plt.xlabel('Consumption ($x$)', fontsize=10)
plt.ylabel(r'Utility ($U(x)$)', fontsize=10)
plt.legend()
plt.grid(color='lightgray')

# 自动调整布局以防止标题或标签被截断
plt.tight_layout()

# 保存图片
output_dir = r'D:\JG\助教推文提交\2020助教推文\_00lian_blogs\Figs'
FigName = 'model_marginal_decrease_01.png'
os.makedirs(output_dir, exist_ok=True)
plt.savefig(os.path.join(output_dir, FigName), dpi=150, bbox_inches='tight')

# 显示图形
plt.show()
```

```python
# Model 2: Cobb-Douglas 生产函数和单因素生产函数

import numpy as np
import os
import matplotlib.pyplot as plt

# 定义 Cobb-Douglas 生产函数的参数
A = 1  # 技术水平
alpha = 0.3  # 资本弹性
beta  = 0.7  # 劳动弹性

# 生成资本 (K) 和劳动 (L) 的网格
K = np.linspace(1, 10, 20)
L = np.linspace(1, 10, 20)
K, L = np.meshgrid(K, L)

# 计算 Cobb-Douglas 生产函数的输出 (Y)
Y = A * (K**alpha) * (L**beta)

# 定义单因素生产函数的参数
K_single = np.linspace(1, 10, 100)
Y_single = A * (K_single**alpha)

fig = plt.figure(figsize=(10, 4))  # 增加整体图片尺寸以加大横向间距

# 左图：Cobb-Douglas 生产函数 3D 图形
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(K, L, Y, cmap='viridis')
ax1.set_title(fr'C-D Function: $Y = AK^\alpha L^\beta$ ($\alpha=${alpha}, $\beta=${beta})', fontsize=10)
ax1.set_xlabel('Capital (K)', fontsize=10)
ax1.set_ylabel('Labor (L)', fontsize=10)
ax1.set_zlabel('Output (Y)', fontsize=10)

# 右图：单因素生产函数 Y = A K^alpha，缩小为 0.7 倍
ax2 = fig.add_axes([0.6, 0.2, 0.28, 0.6])  # 自定义位置和大小，缩小为 0.7 倍
ax2.plot(K_single, Y_single, label=fr'$Y = AK^\alpha$ ($\alpha$={alpha})', color='blue')  
ax2.set_title('Single Input Factor: $Y = AK^\\alpha$', fontsize=10)
ax2.set_xlabel('Capital ($K$)', fontsize=10)
ax2.set_ylabel('Output ($Y$)', fontsize=10)
ax2.grid(color = 'lightgray')
ax2.legend()

# 自动调整布局
plt.tight_layout()

# 保存图片
output_dir = r'D:\JG\助教推文提交\2020助教推文\_00lian_blogs\Figs'
FigName = 'model_marginal_decrease_02.png'
os.makedirs(output_dir, exist_ok=True)
plt.savefig(os.path.join(output_dir, FigName), dpi=150, bbox_inches='tight')

# 显示图形
plt.show()
```

```python
# Model 3: 定义技术进步函数

import numpy as np
import os
import matplotlib.pyplot as plt

# 定义技术进步函数
def tech_progress(t, alpha):
    return t**alpha

# 生成 t 范围
t = np.linspace(1, 5, 100)

# 设置不同的 alpha 值
alpha_values = [0.3, 0.5, 0.8]

# 绘制图形
plt.figure(figsize=(4, 3))  # 设置图片尺寸

for alpha in alpha_values:
    A = tech_progress(t, alpha)
    plt.plot(t, A, label=rf'$\alpha = {alpha}$')

# 设置标题和轴标签，使用 LaTeX 格式
plt.title(r'Technology Progress: $A(t) = t^\alpha$', fontsize=10)
plt.xlabel('Time ($t$)', fontsize=10)
plt.ylabel('Technology Progress ($A(t)$)', fontsize=10)
plt.legend(fontsize=9)
plt.grid(color='lightgray')

# 保存图片
output_dir = r'D:\JG\助教推文提交\2020助教推文\_00lian_blogs\Figs'
FigName = 'model_marginal_decrease_03.png'
os.makedirs(output_dir, exist_ok=True)
plt.savefig(os.path.join(output_dir, FigName), dpi=150, bbox_inches='tight')

# 显示图形
plt.show()
```

```python
# Model 4: Investment Return Model
# This model demonstrates the relationship between investment and return
# using a power function R(I) = I^alpha, where alpha is the return decay factor.

import numpy as np
import os
import matplotlib.pyplot as plt

# 定义投资回报函数
def investment_return(I, alpha):
    return I**alpha

# 生成投资范围
I = np.linspace(1, 10, 100)

# 设置不同的 alpha 值
alpha_values = [0.3, 0.5, 0.7]

# 绘制图形
plt.figure(figsize=(4, 3))  # 设置图片尺寸

for alpha in alpha_values:
    R = investment_return(I, alpha)
    plt.plot(I, R, label=rf'$\alpha = {alpha}$')

# 设置标题和轴标签，使用 LaTeX 格式
plt.title(r'Investment Return: $R(I) = I^\alpha$', fontsize=10)
plt.xlabel('Investment ($I$)', fontsize=10)
plt.ylabel('Return ($R(I)$)', fontsize=10)
plt.legend(fontsize=9)
plt.grid(color='lightgray')

# 保存图片
output_dir = r'D:\JG\助教推文提交\2020助教推文\_00lian_blogs\Figs'
FigName = 'model_marginal_decrease_04.png'
os.makedirs(output_dir, exist_ok=True)
plt.savefig(os.path.join(output_dir, FigName), dpi=180, bbox_inches='tight')

# 显示图形
plt.show()
```

```python
# Model 5: 折现效用函数

import numpy as np
import os
import matplotlib.pyplot as plt

# 定义折现效用函数
def discount_utility(C, beta):
    return np.sum(beta**np.arange(len(C)) * C)

# 生成消费序列
C = np.array([10, 12, 15, 18, 22])
beta_values = [0.8, 0.9, 0.95]

# 创建图形
fig, axes = plt.subplots(1, 2, figsize=(8, 3.5))  # 设置左右图布局和尺寸

# 左图：折现效用函数随时间变化
for beta in beta_values:
    discounted_utilities = [
        discount_utility(C[:t+1], beta) for t in range(len(C))
    ]
    axes[0].plot(
        range(1, len(C)+1), discounted_utilities,
        label=rf'$\beta = {beta}$'
    )

axes[0].set_title(
    r'Discounted Utility Over Time: $U(C_t) = \sum_{t=0}^\infty \beta^t C_t$',
    fontsize=10
)
axes[0].set_xlabel('Time Period (t)', fontsize=9)
axes[0].set_ylabel('Discounted Utility', fontsize=9)
axes[0].set_xticks(range(1, 6))
axes[0].set_xticklabels(['1', '2', '3', '4', '5'])
axes[0].legend(fontsize=9)
axes[0].grid(color='lightgray')
axes[0].text(
    1.1, 65, r'$C_t$: [10, 12, 15, 18, 22]',
    fontsize=9, horizontalalignment='left', verticalalignment='center'
)

# 右图：$\beta^t$ 与 $t$ 的关系
t = np.arange(0, 10)  # 时间范围
for beta in beta_values:
    beta_t = beta**t
    axes[1].plot(t, beta_t, label=rf'$\beta = {beta}$')

axes[1].set_title(
    r'Relationship Between $\beta^t$ and $t$', fontsize=10
)
axes[1].set_xlabel('Time Period (t)', fontsize=9)
axes[1].set_ylabel(r'$\beta^t$', fontsize=9)
axes[1].legend(fontsize=9)
axes[1].grid(color='lightgray')

# 自动调整布局
plt.tight_layout()

# 保存图片
output_dir = r'D:\JG\助教推文提交\2020助教推文\_00lian_blogs\Figs'
FigName = 'model_marginal_decrease_05.png'
os.makedirs(output_dir, exist_ok=True)
plt.savefig(os.path.join(output_dir, FigName), dpi=150, bbox_inches='tight')

# 显示图形
plt.show()
```

```python
# Model 6: 资源消耗函数

import numpy as np
import os
import matplotlib.pyplot as plt

# 定义资源消耗函数
def resource_consumption(s, beta):
    return 1 - np.exp(-beta * s)

# 生成资源消耗范围
s = np.linspace(0, 10, 100)

# 定义三组不同的 beta 值
beta_values = [0.3, 0.5, 0.7]

# 绘制图形
plt.figure(figsize=(4, 3))  # 设置图片尺寸

for beta in beta_values:
    Q = resource_consumption(s, beta)
    plt.plot(s, Q, label=rf'$\beta = {beta}$')

# 设置标题和轴标签
plt.title(r'Resource Consumption: $Q(s) = 1 - e^{-\beta s}$', fontsize=12)
plt.xlabel('Resource Consumption ($s$)', fontsize=9)
plt.ylabel('Effect ($Q(s)$)', fontsize=9)
plt.legend(fontsize=8)
plt.grid(color='lightgray')

# 保存图片
output_dir = r'D:\JG\助教推文提交\2020助教推文\_00lian_blogs\Figs'
FigName = 'model_marginal_decrease_06.png'
os.makedirs(output_dir, exist_ok=True)
plt.savefig(os.path.join(output_dir, FigName), dpi=150, bbox_inches='tight')

# 显示图形
plt.show()
```
