## 图形定制和美化

`matplotlib` 提供了丰富的图形定制选项，以下是一些常见设置：

- **图形尺寸**
  - 设置图形大小：`plt.figure(figsize=(6, 4))`（应最先调用）

- **标题与坐标轴**
  - 设置标题：`plt.title("主标题")`
  - 设置横纵坐标轴标签：`plt.xlabel("X 轴")`、`plt.ylabel("Y 轴")`
  - 设置坐标轴范围：`plt.xlim(0, 10)`、`plt.ylim(-5, 5)`
  - 显示网格线：`plt.grid(True)`

- **图例**
  - 添加图例：`plt.legend()`

- **线条与标记样式**
  - 绘图时设置线型和颜色：`plt.plot(x, y, linestyle='--', color='blue')`
  - 设置标记样式：`marker='o'`、`marker='s'`
  - 设置线宽与透明度：`linewidth=2`、`alpha=0.6`

- **字体与颜色**
  - 全局字体设置：`plt.rcParams['font.size'] = 12`
  - 设置背景颜色：`plt.gcf().set_facecolor('white')`
  - 设置边框与坐标轴颜色：
    - `plt.gca().spines['top'].set_color('gray')`
    - `plt.tick_params(axis='x', colors='red', length=8)`

### 实例 1：绘制折线图并设置标题和坐标轴标签

本例中， 我们将使用上面介绍的选项，来美化此前绘制的折线图。有如下几点说明：
- 中文支持。若标题或坐标轴标签中包含中文字符，可能会出现乱码。可以使用 `plt.rcParams['font.sans-serif'] = ['SimHei']` 来设置字体为黑体，或者使用 `font_manager` 模块来加载自定义字体。
- 保存图片。`plt.savefig()` 可以将图形保存为文件，支持多种格式，也可以设定图片尺寸。需要注意的是，该命令需要在 `plt.show()` 之前调用，否则输出的图片会是空白的。

````python
import matplotlib.pyplot as plt

# 设置中文字体支持（如已设置可省略）
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 数据
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# 简单的折线图
plt.figure(figsize=(5, 3))  # 设置图形大小
plt.plot(x, y, marker='o', markerfacecolor='red')
plt.title("简单折线图")
plt.xlabel("广告投入 (万元)", fontsize=10)
plt.ylabel("销售额 (万元)", fontsize=10)
plt.xticks(range(min(x), max(x) + 1, 1))  # 设置 x 轴刻度步长为 1
plt.yticks(range(0, max(y) + 1, 5))       # 设置 y 轴刻度步长为 5
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.annotate("数据来源：时尚未来公司", 
             xy=(-0.0, -0.3), 
             xycoords='axes fraction', 
             ha='left', 
             fontsize=9)
# 保存图形
plt.savefig("simple_line_chart.png", dpi=300, bbox_inches='tight')
# 显示图形：一定要放在最后，因为执行 plt.show()后，绘图区会被清空
plt.show()
````

## 图形的基本组成


matplotlib 的绘图结构包括以下几个基本要素：

- figure：图形窗口
- axes：图像区域（包含坐标轴）
- title：图形标题
- xlabel, ylabel：坐标轴标签
- legend：图例
- grid：网格线

这些元素共同构成一张完整的图。

## 子图

我们也可以在同一张图中绘制多个子图。使用 `plt.subplot()` 函数可以创建一个包含多个子图的图形。

扩展阅读：

- [matplotlib - Subplots](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplot_demo.html)
- [Arranging multiple Axes in a Figure](https://matplotlib.org/stable/users/explain/axes/arranging_axes.html#arranging-axes)

下面是一个示例：

````python
# 创建 1 行 3 列的子图
fig, subfig = plt.subplots(1, 3, figsize=(5, 2))  

# 在第一个子图中绘制散点图
subfig[0].scatter(x, y)
subfig[0].set_title("散点图")

# 在第二个子图中绘制柱状图
subfig[1].bar(x, y)
subfig[1].set_title("柱状图")

# 在第三个子图中绘制面积图
subfig[2].fill_between(x, y, color='skyblue', alpha=0.5)
subfig[2].set_title("面积图")

plt.tight_layout()  # 调整布局以防止重叠
plt.savefig(r"figs\subplots_03.png", dpi=100, bbox_inches='tight')  # 保存图形
plt.show()
````

### 解读：元组解包语法

在上述代码中，`fig, subfig = plt.subplots(1, 3, figsize=(5, 2))` 这一行初看语法似乎不太直观，其实是 `matplotlib` 中的常见用法，结合了 **函数的多值返回机制** 和 **元组解包（tuple unpacking）** 的语法结构，现作如下说明。

这一行代码的作用是：**同时创建一个图形对象（Figure）和三个子图（Axes）对象**。该函数返回一个包含两个元素的 **元组**（tuple）：

* 第一个元素是 `Figure` 对象，代表整张图；
* 第二个元素是一个 `Axes` 对象的数组，每个元素都是一个子图。

由于 Python 支持对元组进行解包，我们可以使用如下方式将这两个元素分别赋值给两个变量 `fig` 和 `subfig`：

```python
fig, subfig = plt.subplots(1, 3, figsize=(5, 2))
```

这种语法称为「元组解包（tuple unpacking）」，是 Python 中的一种常见语法结构。它允许我们将一个返回多个值的函数，直接对应赋值给多个变量。

#### 元组解包实例

```python
a, b = (1, 2)  # 相当于 a = 1, b = 2
```

或者：

```python
def get_name_age():
    return ("Alice", 25)

name, age = get_name_age()
```

在 `plt.subplots()` 中也是同理，其返回值是一个形如 `(fig, subfig)` 的元组，因此可以直接用 `fig, subfig = ...` 来解包。


#### 参数和返回值说明

`plt.subplots(1, 3, figsize=(5, 2))` 中的参数说明如下：

* `1` 表示生成 **1 行**子图；
* `3` 表示生成 **3 列**子图；
* `figsize=(5, 2)` 设置整张图的尺寸为宽 5 英寸、高 2 英寸。

`fig, subfig = plt.subplots(...)` 的返回值说明如下：

* `fig` 是一个 `matplotlib.figure.Figure` 对象，表示整张图；
* `subfig` 是一个包含 3 个子图的数组，类型为 `numpy.ndarray`，可通过索引 `subfig[0]`、`subfig[1]` 和 `subfig[2]` 分别访问每个子图。

如果我们只生成一个子图（例如 `plt.subplots(1, 1)`），那么 `subfig` 就是一个单独的 `Axes` 对象，而不是数组。

#### 等价写法（不推荐）

为了帮助理解，我们也可以将 `plt.subplots()` 的返回值用更繁琐的方式手动拆解，虽然不推荐实际使用：

```python
result = plt.subplots(1, 3, figsize=(5, 2))
fig = result[0]
subfig = result[1]
```

推荐使用解包形式 `fig, subfig = ...`，更简洁、直观，是 `matplotlib` 多子图操作中的标准写法。
