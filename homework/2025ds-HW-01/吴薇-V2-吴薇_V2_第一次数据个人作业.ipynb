## 开始作业
   * 添加一个 Python 代码单元格，根据如下提示词生成 Python 代码，产生模拟数据 （你可以使用 Copilot 或 ChatGPT 等工具）生成模拟数据：
     * - 1.N = 500，包含 2 个变量 (x1, x2)，x1 ~ N(0, 1)，x2 ~ N(3, 1.5)，corr(x1, x2) = 0.4。
     * - 2. 存入一个名为 df 的数据框中。
     * - 3. 种子值为：'171'。

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 设置中文字体（如黑体）
plt.rcParams['font.family'] = ['SimHei', 'DejaVu Sans']  # 优先使用 SimHei，缺失字符时用 DejaVu Sans 补全
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题（用 DejaVu Sans 的负号）

# 设置随机种子为171
np.random.seed(171)

N = 500
mean = [0, 3]
std_x1 = 1
std_x2 = 1.5
corr = 0.4
cov = corr * std_x1 * std_x2  # 协方差

# 协方差矩阵
cov_matrix = [[std_x1**2, cov],
              [cov, std_x2**2]]

# 生成多元正态分布数据
x1, x2 = np.random.multivariate_normal(mean, cov_matrix, N).T

# 存入数据框
df = pd.DataFrame({'x1': x1, 'x2': x2})

# 设置显示格式，保留两位小数
pd.set_option('display.float_format', lambda x: f'{x:.2f}')

# 查看数据的基本统计信息
print('数据基本统计信息：')
print(df.describe())
```

## 任务5：在同一幅图中绘制 x1 和 x2 的箱线图，了解上须线、下须线、异常值、四分位数等情况。

```python
# 任务5：在同一幅图中绘制 x1 和 x2 的箱线图
import seaborn as sns
import matplotlib.pyplot as plt
# 设置中文字体（如黑体）
plt.rcParams['font.family'] = ['SimHei', 'DejaVu Sans']  # 优先使用 SimHei，缺失字符时用 DejaVu Sans 补全
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题（用 DejaVu Sans 的负号）

plt.figure(figsize=(8, 6))
ax = sns.boxplot(data=df[['x1', 'x2']])

# 计算中位数
med_x1 = df['x1'].median()
med_x2 = df['x2'].median()

# 标注中位数
plt.text(0, med_x1, f'{med_x1:.2f}', ha='center', va='bottom', color='blue', fontsize=10)
plt.text(1, med_x2, f'{med_x2:.2f}', ha='center', va='bottom', color='orange', fontsize=10)

# 颜色定义
COLOR_Q1 = '#3498db'    # 蓝色 - 下四分位数
COLOR_Q3 = '#e74c3c'    # 红色 - 上四分位数
COLOR_OUTLIER = '#d35400'  # 橙色 - 异常值
COLOR_UPPER_BOUND = '#2ecc71'  # 绿色 - 上边界
COLOR_LOWER_BOUND = '#9b59b6'  # 紫色 - 下边界

# 标注异常值、四分位数和边界
for i, col in enumerate(['x1', 'x2']):
    # 计算统计量
    q1 = df[col].quantile(0.25)  # 下四分位数
    q3 = df[col].quantile(0.75)  # 上四分位数
    iqr = q3 - q1               # 四分位距
    
    # 计算上下边界
    upper_bound = q3 + 1.5 * iqr  # 上虚线位置
    lower_bound = q1 - 1.5 * iqr  # 下虚线位置
    
    # 手动计算 MAD（平均绝对偏差）
    median_val = df[col].median()
    mad = (df[col] - median_val).abs().median()
    
    # 使用四分位距方法识别异常值
    outliers = df[(df[col] < lower_bound) | 
                  (df[col] > upper_bound)][col]
    
    # 绘制上四分位数和下四分位数的水平线
    plt.axhline(y=q1, xmin=i-0.45, xmax=i-0.35, color=COLOR_Q1, linestyle='--', alpha=0.7, linewidth=1.5)
    plt.axhline(y=q3, xmin=i-0.45, xmax=i-0.35, color=COLOR_Q3, linestyle='--', alpha=0.7, linewidth=1.5)
    
    # 绘制上下边界虚线
    plt.axhline(y=upper_bound, xmin=i-0.45, xmax=i-0.35, color=COLOR_UPPER_BOUND, linestyle=':', alpha=0.7, linewidth=1.5)
    plt.axhline(y=lower_bound, xmin=i-0.45, xmax=i-0.35, color=COLOR_LOWER_BOUND, linestyle=':', alpha=0.7, linewidth=1.5)
    
    # 标注上四分位数和下四分位数的值（位置微调）
    plt.text(i-0.2, q1, f'Q1: {q1:.2f}', ha='right', va='center', 
             color=COLOR_Q1, fontsize=9, bbox=dict(facecolor='white', alpha=0.7))
    plt.text(i-0.2, q3, f'Q3: {q3:.2f}', ha='right', va='center', 
             color=COLOR_Q3, fontsize=9, bbox=dict(facecolor='white', alpha=0.7))
    
    # 标注上下边界的值（更靠左）
    plt.text(i-0.3, upper_bound, f'上限: {upper_bound:.2f}', ha='right', va='center', 
             color=COLOR_UPPER_BOUND, fontsize=8, bbox=dict(facecolor='white', alpha=0.7))
    plt.text(i-0.3, lower_bound, f'下限: {lower_bound:.2f}', ha='right', va='center', 
             color=COLOR_LOWER_BOUND, fontsize=8, bbox=dict(facecolor='white', alpha=0.7))
    
    # 智能标注异常值
    if col == 'x2':
        # 将异常值分为上部和下部
        upper_outliers = outliers[outliers >= q3]
        lower_outliers = outliers[outliers < q3]
        
        # 处理上部异常值（简单垂直偏移）
        for y in upper_outliers:
            plt.scatter(i, y, color=COLOR_OUTLIER, marker='o', s=30, edgecolors='k', zorder=5)
            plt.annotate(f'{y:.2f}',
                         xy=(i, y),
                         xytext=(5, -10),  # 固定向下偏移
                         textcoords='offset points',
                         fontsize=9,
                         color='darkred',
                         bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.7),
                         va='top')
        
        # 处理下部异常值（网格布局）
        if not lower_outliers.empty:
            # 对下部异常值进行排序
            sorted_lower = sorted(lower_outliers)
            
            # 计算网格参数
            n = len(sorted_lower)
            cols = min(3, n)  # 最多3列
            rows = (n + cols - 1) // cols  # 向上取整
            
            # 创建网格位置
            grid_positions = []
            for row in range(rows):
                for col_idx in range(cols):
                    if len(grid_positions) < n:
                        # 计算网格中的相对位置
                        x_offset = 10 + col_idx * 40  # 列间距
                        y_offset = 10 + row * 15      # 行间距
                        grid_positions.append((x_offset, y_offset))
            
            # 为每个异常值分配网格位置
            for y, (x_off, y_off) in zip(sorted_lower, grid_positions):
                plt.scatter(i, y, color=COLOR_OUTLIER, marker='o', s=30, edgecolors='k', zorder=5)
                plt.annotate(f'{y:.2f}',
                             xy=(i, y),
                             xytext=(x_off, y_off),
                             textcoords='offset points',
                             fontsize=9,
                             color='darkred',
                             bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.7),
                             va='bottom',
                             ha='left')
    else:  # 对x1使用常规标注
        for y in outliers:
            plt.scatter(i, y, color=COLOR_OUTLIER, marker='o', s=30, edgecolors='k', zorder=5)
            plt.annotate(f'{y:.2f}',
                         xy=(i, y),
                         xytext=(5, 5),
                         textcoords='offset points',
                         fontsize=9,
                         color='darkred',
                         bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.7))
plt.title('x1 和 x2 的箱线图（含中位数和异常值标注）')
plt.xlabel('变量')
plt.ylabel('值')
```

```python
# 任务8：绘制 x1 和 x2 的散点图，添加回归线和置信区间
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.regplot(x='x1', y='x2', data=df, ci=95, scatter_kws={'color': '#a6cee3'}, line_kws={'color': 'red'})
plt.title('x1 和 x2 的散点图及回归线')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

### 分析：
   **散点图解读**
   * *散点图：展示了 x1 和 x2 的散点分布情况。每个点代表一组生成值，x1 和 x2 的取值组合在图中形成的点。*
   * *横轴（x1）：代表变量 x1 ，取值范围大致从 -3 到 3 ，是自变量所在轴。*
   * *纵轴（x2）：代表变量 x2 ，取值范围大致从 0 到 6 ，是因变量所在轴 。*
   * *散点：图中蓝色的点为散点，每个点对应一组 x1 和 x2 的观测值。散点分布显示了数据的实际观测情况，体现了两个变量在不同取值组合下的样本点位置。*
   * *回归线：红色直线是回归线，它基于一定的统计方法（如最小二乘法）拟合得出，用于描述 x1 和 x2 之间的线性关系趋势，直观展示变量间平均变化趋势。*
   * *阴影区域：回归线周围的红色阴影区域通常表示预测区间或置信区间。它反映了在给定 x1 值时，对 x2 预测值的不确定性范围。在置信区间内，我们有一定的概率认为真实的 x2 值会落在其中。*
   
   **变量关系解读**
   * *线性关系：从图中可以看出，散点大致围绕回归线分布，说明 x1 和 x2 之间存在线性相关关系。回归线斜率为正，表明 x1 与 x2 呈正相关，即 x1 增大时，x2 倾向于增大 。*
   * *离散程度：散点并非完全落在回归线上，说明 x1 和 x2 之间不是确定性的函数关系，存在一定的随机波动。散点分布越分散，说明变量关系的不确定性越大；越集中在回归线附近，则变量间的线性关系越紧密。*
