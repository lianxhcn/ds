```python
import numpy as np
import pandas as pd

# 设置随机种子
np.random.seed(157)

# 定义变量数量和样本量
n = 500

# 创建相关系数矩阵
corr_matrix = np.array([
    [1.0, 0.4],
    [0.4, 1.0]
])

# 创建标准差数组
std_devs = np.array([1.0, 1.5])

# 计算协方差矩阵
cov_matrix = np.outer(std_devs, std_devs) * corr_matrix

# 生成多元正态分布数据
mean = [0, 3]  # x1均值为0，x2均值为3
data = np.random.multivariate_normal(mean, cov_matrix, size=n)

# 创建DataFrame
df = pd.DataFrame(data, columns=['x1', 'x2'])

print("数据基本统计量：")
print(df.describe())   
```

任务一：呈现 df 数据框的前 5 行。

```python
print("\n数据前5行内容：")
print(df.head().to_string())
```

# **结果说明**：
# 数据框 `df` 包含 500 行，前 5 行显示了 `x1` 和 `x2` 的部分数值，数值范围大致符合正态分布。

# 任务二：呈现 df 数据框的描述性统计，包括均值，标准差，中位数，最大值，最小值，偏度，峰度。

```python
# 呈现完整描述性统计
print("\n完整描述性统计：")
stats = pd.DataFrame({
    '均值': df.mean(),
    '标准差': df.std(),
    '中位数': df.median(),
    '最大值': df.max(),
    '最小值': df.min(),
    '偏度': df.skew(),
    '峰度': df.kurt()
})

print(stats.round(4))
```

# 任务二结果说明：  
- `x1` 的均值接近 0，标准差接近 1（符合 N(0,1)）。  
- `x2` 的均值接近 3，标准差接近 1.5（符合 N(3,1.5)）。  
- 偏度（Skewness）和峰度（Kurtosis）接近正态分布的 0 和 3。

```python

# 在同一幅图中绘制x1和x2的密度函数图
plt.figure(figsize=(12, 7))

# 绘制x1的核密度估计曲线
sns.kdeplot(df['x1'], shade=True, color='blue', alpha=0.3, label='x1 密度')

# 绘制x2的核密度估计曲线
sns.kdeplot(df['x2'], shade=True, color='orange', alpha=0.3, label='x2 密度')

# 添加垂直线标记均值位置
plt.axvline(x=df['x1'].mean(), color='blue', linestyle='--', alpha=0.7)
plt.axvline(x=df['x2'].mean(), color='orange', linestyle='--', alpha=0.7)

# 添加标题和标签
plt.title('x1与x2的密度函数对比图', fontsize=15)
plt.xlabel('变量值', fontsize=13)
plt.ylabel('密度', fontsize=13)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

# 添加文本标注均值
plt.text(df['x1'].mean() + 0.1, 0.1, f"x1均值: {df['x1'].mean():.2f}", color='blue')
plt.text(df['x2'].mean() + 0.1, 0.1, f"x2均值: {df['x2'].mean():.2f}", color='orange')

plt.tight_layout()
plt.show()
```

任务七：计算 df 中 x1 和 x2 的相关系数，并输出结果。

```python
# 新增：计算x1和x2的相关系数并输出结果
# 计算皮尔逊相关系数矩阵
correlation_matrix = df.corr()

# 获取x1和x2之间的相关系数
pearson_corr = correlation_matrix.loc['x1', 'x2']

# 计算斯皮尔曼相关系数
spearman_corr = df['x1'].corr(df['x2'], method='spearman')

# 计算肯德尔相关系数
kendall_corr = df['x1'].corr(df['x2'], method='kendall')

print("\n相关系数分析结果：")
print(f"x1和x2的皮尔逊相关系数: {pearson_corr:.4f}")
print(f"x1和x2的斯皮尔曼相关系数: {spearman_corr:.4f}")
print(f"x1和x2的肯德尔相关系数: {kendall_corr:.4f}")

# 可视化相关系数矩阵
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True)
plt.title('x1和x2的相关系数矩阵热力图', fontsize=14)
plt.tight_layout()
plt.show()    
```

# 任务七运行结果说明：
# 计算结果约为 0.375，与模拟数据设定的 `corr(x1,x2)=0.4` 接近，表明两者存在中等程度的正相关。

任务八：绘制 x1 和 x2 的散点图，添加回归线和置信区间。

```python
# 新增：绘制x1和x2的散点图，添加回归线和置信区间
plt.figure(figsize=(12, 8))

# 使用seaborn的regplot绘制散点图和回归线
sns.regplot(x='x1', y='x2', data=df, 
            scatter_kws={'alpha': 0.5, 'color': 'blue', 's': 40},
            line_kws={'color': 'red', 'lw': 2.5})

# 添加标题和标签
plt.title('x1与x2的散点图及线性回归', fontsize=16)
plt.xlabel('x1值', fontsize=14)
plt.ylabel('x2值', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# 添加网格线
plt.grid(True, linestyle='--', alpha=0.6)

# 计算并显示回归方程和相关系数
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(df['x1'], df['x2'])

# 在图上添加回归方程和相关系数信息
plt.annotate(f'回归方程: y = {slope:.3f}x + {intercept:.3f}\n'
             f'皮尔逊相关系数 (r): {pearson_corr:.3f}\n'
             f'决定系数 (R²): {r_value**2:.3f}\n'
             f'p值: {p_value:.3e}',
             xy=(0.05, 0.95), xycoords='axes fraction',
             bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8),
             fontsize=12, ha='left', va='top')

# 调整布局
plt.tight_layout()
plt.show()    
```

# 任务八运行结果说明：
# 散点分布：由于 x1 和 x2 存在正相关关系（相关系数约为 0.4），散点会大致呈现从左下到右上的分布趋势。
# 回归线：红色直线是根据数据拟合的回归线，表示 x1 和 x2 之间的最佳线性关系。斜率表示 x1与X2的相关变量。
# 置信区间：回归线周围的浅红色区域是 95% 置信区间，表示真实的回归线有 95% 的可能性落在这个区域内。置信区间越窄，说明回归模型的可靠性越高。
# 相关系数：图的左上角标注了 x1 和 x2 的相关系数r（约为 0.4）
