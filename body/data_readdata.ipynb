````python
def myrepeat(str, times):
    """
    重复字符串指定次数
    """
    if times < 0:
        print("error: times must be non-negative value")
        exit(1)
    elif times == 0:
        return ""
    else:
        return str * times

myrepeat("lianxh", 2)
myrepeat("dog", -2)
````

编写上述函数时，初学者容易犯的错误是：
- 第一行的行尾忘记加上冒号
- Python 中的 `elif` 等价于 `else if`，这与 R 和 Stata 不同。
